# Análise Exploratória de Dados de Vendas 

```
# Importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")

# seaborn => estilo do gráfico no matplotlib (há também uma biblioteca com este nome)
```

```
# Upload do arquivo (somente quando executado no Google Colaboratory)
# from google.colab import files
# arq = files.upload()
```

```
# Criando nosso DataFrame
df = pd.read_excel("datasets/AdventureWorks.xlsx")
```

```
# Visualizando as 5 primeiras linhas
df.head()
```

```
# Quantidade de linhas e colunas
df.shape
```

```
# Verificando os tipos de dados
df.dtypes
```

```
# Qual a Receita total?
df["Valor Venda"].sum()

# "Valor Venda" => valor final de venda
```

```
# Criação da coluna "custo" (custo total)
df["custo"] = df["Custo Unitário"].mul(df["Quantidade"]) 
```

```
df.head(1)
```

```
# Qual o custo Total?
round(df["custo"].sum(), 2)
```

```
# Criação da coula "lucro" (lucro total)
df["lucro"]  = df["Valor Venda"] - df["custo"] 
```

```
df.head(1)
```

```
# Qual o lucro Total?
round(df["lucro"].sum(),2)
```

```
# Criação de uma coluna com total de dias para enviar o produto
df["Tempo_envio"] = df["Data Envio"] - df["Data Venda"]
```

```
df.head(1)
```

**Agora, queremos saber a média do tempo de envio para cada Marca, e para isso precisamos transformar a coluna Tempo_envio em númerica**

```
# Extraindo apenas os dias
df["Tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```
df.head(1)
```

```
# Verificando o tipo da coluna Tempo_envio
df["Tempo_envio"].dtype
```

```
# Média do tempo de envio por Marca
df.groupby("Marca")["Tempo_envio"].mean()
```

# **Missing Values**

```
# Verificando se temos dados faltantes
df.isnull().sum()
```

**E, se a gente quiser saber o Lucro por Ano e Por Marca?**

```
# Agrupando os dados por ano e marca
df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum()
```

```
 # Configurando o formato dos números decimais do lucro por marca
 pd.options.display.float_format = '{:20,.2f}'.format
```

```
# Resetando o index
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum().reset_index()
lucro_ano

# "reset_index()" => resseta o índice e retorna na forma de colunas
```

```
# Qual o total de produtos vendidos (ordem decrescente de quantidade vendida)?
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```
# Gráfico Total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total Produtos Vendidos")
plt.xlabel("Total")
plt.ylabel("Produto");
```

```
df.groupby(df["Data Venda"].dt.year)["lucro"].sum().plot.bar(title="Lucro x Ano")
plt.xlabel("Ano")
plt.ylabel("Receita");
```

```
df.groupby(df["Data Venda"].dt.year)["lucro"].sum()
```

```
# Selecionando apenas as vendas de 2009
df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```
df_2009.head()
```

```
df_2009.groupby(df_2009["Data Venda"].dt.month)["lucro"].sum().plot(title="Lucro x Mês")
plt.xlabel("Mês")
plt.ylabel("Lucro");
```

```
df_2009.groupby("Marca")["lucro"].sum().plot.bar(title="Lucro x Marca")
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal');
```

```
df_2009.groupby("Classe")["lucro"].sum().plot.bar(title="Lucro x Classe")
plt.xlabel("Classe")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal');
```

```
# Análise estatística (variável "Tempo_envio")
df["Tempo_envio"].describe()
```

```
# Moda da variável "Tempo_envio"
df["Tempo_envio"].mode()
```

```
# Gráfico de Boxplot
plt.boxplot(df["Tempo_envio"]);
plt.title("Boxplot - Tempo de envio")
```
